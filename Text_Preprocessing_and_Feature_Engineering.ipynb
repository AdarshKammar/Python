```python
import nltk
import string
import numpy as np
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from collections import Counter

# Download necessary resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Sample text dataset
text = "Natural Language Processing is an interesting field of AI. It enables machines to understand and process human language."

# 1. Stopwords Removal
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    filtered_text = [word for word in words if word not in stop_words and word not in string.punctuation]
    return filtered_text

filtered_tokens = remove_stopwords(text)
print("Filtered Tokens:", filtered_tokens)

# 2. Tokenization
words = word_tokenize(text)
sentences = sent_tokenize(text)
print("Word Tokens:", words)
print("Sentence Tokens:", sentences)

# 3. Stemming and Lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Stemmed Words:", stemmed_words)
print("Lemmatized Words:", lemmatized_words)

# 4. POS Tagging
pos_tags = pos_tag(filtered_tokens)
print("POS Tags:", pos_tags)

# 5. TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([text])
print("TF-IDF Matrix:", tfidf_matrix.toarray())

# 6. One-Hot Encoding
def one_hot_encoding(text):
    words = set(word_tokenize(text.lower()))
    vocab = {word: i for i, word in enumerate(words)}
    one_hot_vectors = {word: np.eye(len(vocab))[i] for word, i in vocab.items()}
    return one_hot_vectors

one_hot_vectors = one_hot_encoding(text)
print("One-Hot Encodings:", one_hot_vectors)

# 7. Bag of Words
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform([text])
print("Bag of Words:", bow_matrix.toarray())

# 8. N-Gram Representations
def generate_ngrams(text, n):
    words = word_tokenize(text.lower())
    ngrams = [tuple(words[i:i+n]) for i in range(len(words)-n+1)]
    return ngrams

unigrams = generate_ngrams(text, 1)
bigrams = generate_ngrams(text, 2)
trigrams = generate_ngrams(text, 3)

print("Unigrams:", unigrams)
print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
```
